```python
CLO_THRESHOLD = 50      # CLO attainment threshold (%)
PASS_MARK     = 60      # Minimum % to pass
TOP_N         = 10      # Top / Bottom N students

# Colour palette
C = dict(
    primary="#1a73e8", success="#34a853", danger="#ea4335",
    warning="#fbbc04", info="#46bdc6",   purple="#9c27b0",
    bg="#ffffff",      card="#f8f9fa",
)

GRADE_COLORS = dict(
    A="#1565c0", B_plus="#2e7d32", B="#388e3c",
    C_plus="#f57f17", C="#f9a825", F="#b71c1c",
)

def grade_color(g):
    mapping = {
        "A": "#1565c0", "A-": "#1976d2",
        "B+": "#2e7d32", "B": "#388e3c", "B-": "#43a047",
        "C+": "#f57f17", "C": "#f9a825", "C-": "#fbc02d",
        "D":  "#e65100", "F": "#b71c1c",
    }
    return mapping.get(str(g).strip(), "#90a4ae")

print(f"⚙️  CLO threshold={CLO_THRESHOLD}%  |  Pass mark={PASS_MARK}%  |  Top N={TOP_N}")
```

## Step 4 — Extract & Clean Student Data

```python
def safe_float(val, default=0.0):
    """Convert cell value to float; return default on None / error strings."""
    if val is None:
        return default
    try:
        f = float(val)
        return f if not np.isnan(f) else default
    except (ValueError, TypeError):
        return default

wb = openpyxl.load_workbook(EXCEL_PATH, data_only=True)

# ── Read course metadata ──
ws_main = wb["Final Combined Marks Sheet"]
meta = {}
for row in ws_main.iter_rows(min_row=1, max_row=10, values_only=True):
    if row[0] and row[2]:
        meta[str(row[0]).strip()] = str(row[2]).strip()

COURSE_TITLE = meta.get("Course Title",   "Introduction to Artificial Intelligence")
COURSE_CODE  = meta.get("Course Code",    "AIC-211")
FACULTY      = meta.get("Name of Faculty","Mr. Abdul Baqi Malik")
SEMESTER     = meta.get("Offered in",     "Fall 2025")
PROGRAM      = meta.get("Program",        "Artificial Intelligence")

# ── Student rows (row 12 onwards) ──
students = []
for row in ws_main.iter_rows(min_row=12, max_row=500, values_only=True):
    sr = row[0]
    if sr is None or not isinstance(sr, (int, float)):
        continue
    reg  = str(row[1]).strip() if row[1] else ""
    name = str(row[2]).strip() if row[2] else ""
    if not reg or not name:
        continue
    students.append({
        "Sr"          : int(sr),
        "Reg_No"      : reg,
        "Name"        : name,
        "Assignment"  : safe_float(row[3]),
        "Quiz"        : safe_float(row[4]),
        "Presentation": safe_float(row[5]),
        "MidTerm"     : safe_float(row[6]),
        "EndTerm"     : safe_float(row[7]),
        "Weighted"    : safe_float(row[8]),
        "TotalMarks"  : safe_float(row[9]),
        "Grade"       : str(row[10]).strip() if row[10] else "F",
        "Remarks"     : str(row[11]).strip() if row[11] and str(row[11]) != "None" else "",
    })

df = pd.DataFrame(students)
df["Percentage"] = df["TotalMarks"].round(2)
df["Pass_Fail"]  = df["Percentage"].apply(lambda x: "Pass" if x >= PASS_MARK else "Fail")

# ── CLO data from 'Quantized Result' ──
ws_clo = wb["Quantized Result"]

# Auto-detect CLO column positions from header row 11
clo_headers = []
for idx, val in enumerate(
    list(ws_clo.iter_rows(min_row=11, max_row=11, values_only=True))[0][3:],
    start=3
):
    if val and str(val).strip().upper().startswith("CLO"):
        clo_headers.append((idx, str(val).strip()))

CLO_NAMES = [c[1] for c in clo_headers]   # e.g. ['CLO 1', 'CLO 2', 'CLO 3']

clo_rows = []
for row in ws_clo.iter_rows(min_row=13, max_row=500, values_only=True):
    if row[0] is None or not isinstance(row[0], (int, float)):
        continue
    reg  = str(row[1]).strip() if row[1] else ""
    name = str(row[2]).strip() if row[2] else ""
    if not reg:
        continue
    entry = {"Reg_No": reg, "Name": name}
    for col_idx, clo_name in clo_headers:
        entry[clo_name] = safe_float(row[col_idx])
    clo_rows.append(entry)

df_clo = pd.DataFrame(clo_rows)
df     = df.merge(df_clo[["Reg_No"] + CLO_NAMES], on="Reg_No", how="left")

# Per-student CLO pass/fail status
for clo in CLO_NAMES:
    df[f"{clo}_Status"] = df[clo].apply(
        lambda x: "Pass" if pd.notna(x) and x >= CLO_THRESHOLD else "Fail"
    )

wb.close()

print(f"\n✅ Extracted {len(df)} students | CLOs: {CLO_NAMES}")
print(f"   {COURSE_CODE}  {COURSE_TITLE}")
print(f"   Faculty: {FACULTY}  |  Semester: {SEMESTER}")
display(df[["Sr","Reg_No","Name","Percentage","Grade","Pass_Fail"] + CLO_NAMES].head(10))
```

## Step 5 — Compute Summary Statistics

```python
TOTAL   = len(df)
PASSED  = int((df["Pass_Fail"] == "Pass").sum())
FAILED  = TOTAL - PASSED
PASS_PC = round(PASSED / TOTAL * 100, 1)
FAIL_PC = round(100 - PASS_PC, 1)
AVG     = round(df["Percentage"].mean(), 2)
HIGH    = round(df["Percentage"].max(),  2)
LOW     = round(df["Percentage"].min(),  2)
MEDIAN  = round(df["Percentage"].median(), 2)
STD     = round(df["Percentage"].std(),  2)

GRADE_ORDER = ["A","A-","B+","B","B-","C+","C","C-","D","F"]
grade_dist  = df["Grade"].value_counts().reindex(GRADE_ORDER).dropna().astype(int)

ASSG_STATS = {
    "Assignment (10)"   : (df["Assignment"].mean(),   10),
    "Quiz (10)"         : (df["Quiz"].mean(),          10),
    "Presentation (15)" : (df["Presentation"].mean(),  15),
    "Mid Term (25)"     : (df["MidTerm"].mean(),       25),
    "End Term (40)"     : (df["EndTerm"].mean(),       40),
}

CLO_STATS = {}
for clo in CLO_NAMES:
    vals          = df[clo].dropna()
    pass_cnt      = int((df[f"{clo}_Status"] == "Pass").sum())
    fail_cnt      = int((df[f"{clo}_Status"] == "Fail").sum())
    CLO_STATS[clo] = {
        "avg"       : round(vals.mean(), 1),
        "pass_cnt"  : pass_cnt,
        "fail_cnt"  : fail_cnt,
        "pass_pct"  : round(pass_cnt / TOTAL * 100, 1),
    }

# Console summary
print("\n" + "=" * 55)
print("  CLASS PERFORMANCE SUMMARY")
print("=" * 55)
print(f"  Total Students : {TOTAL}")
print(f"  Passed         : {PASSED}  ({PASS_PC}%)")
print(f"  Failed         : {FAILED}  ({FAIL_PC}%)")
print(f"  Average        : {AVG}%")
print(f"  Highest        : {HIGH}%")
print(f"  Lowest         : {LOW}%")
print(f"  Median         : {MEDIAN}%   Std Dev: {STD}")
print()
print("  GRADE DISTRIBUTION")
print("  " + "-" * 32)
for g, cnt in grade_dist.items():
    print(f"  {g:3s} | {'█' * cnt} {cnt}")
print()
print("  CLO ATTAINMENT")
print("  " + "-" * 48)
for clo, s in CLO_STATS.items():
    flag = "✅" if s["avg"] >= CLO_THRESHOLD else "⚠️  BELOW THRESHOLD"
    print(f"  {clo}: Avg={s['avg']}%  Pass={s['pass_cnt']}  Fail={s['fail_cnt']}  {flag}")
```

```python
comp_names = list(ASSG_STATS.keys())
comp_avg   = [v[0] for v in ASSG_STATS.values()]
comp_max   = [v[1] for v in ASSG_STATS.values()]
comp_pct   = [round(a/m*100, 1) for a, m in zip(comp_avg, comp_max)]
pct_colors = [C["success"] if p>=60 else C["warning"] if p>=40 else C["danger"] for p in comp_pct]

fig7 = make_subplots(
    rows=1, cols=2,
    subplot_titles=["Avg Obtained vs Max Marks", "Performance % per Component"],
)
fig7.add_trace(go.Bar(
    name="Max Marks", x=comp_names, y=comp_max,
    marker_color="#cfd8dc",
    hovertemplate="%{x}<br>Max: %{y}<extra></extra>",
), row=1, col=1)
fig7.add_trace(go.Bar(
    name="Avg Obtained", x=comp_names, y=[round(v,2) for v in comp_avg],
    marker_color=C["primary"],
    text=[round(v,2) for v in comp_avg], textposition="outside",
    hovertemplate="%{x}<br>Avg: %{y:.2f}<extra></extra>",
), row=1, col=1)
fig7.add_trace(go.Bar(
    name="% Score", x=comp_names, y=comp_pct,
    marker_color=pct_colors,
    text=[f"{p}%" for p in comp_pct], textposition="outside",
    showlegend=False,
    hovertemplate="%{x}<br>Avg %%: %{y:.1f}%%<extra></extra>",
), row=1, col=2)
fig7.update_layout(
    title="<b>Assessment Component Performance</b>",
    barmode="group", plot_bgcolor="#f9f9f9", paper_bgcolor=C["bg"],
    height=430, legend=dict(orientation="h", y=1.1),
    margin=dict(t=80, b=40),
)
fig7.update_yaxes(gridcolor="#e0e0e0")
fig7.show()
```
